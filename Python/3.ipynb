```python
import csv

# Open the CSV file
with open('a.csv', 'r') as csvfile:
    reader = csv.DictReader(csvfile)
    
    # Set the column names to check for common elements
    col1_name = 'Present'
    col2_name = 'Total'
    
    # Create sets to store the values from each column
    col1_values = set()
    col2_values = set()
    
    # Loop through each row in the CSV file
    for row in reader:
        # Add the values from each column to their respective sets
        col1_values.add(row[col1_name])
        col2_values.add(row[col2_name])
    
    # Find the intersection of the two sets to get the common elements
    common_elements = col1_values.intersection(col2_values)
    
    # Print the common elements
    print(common_elements)
```
